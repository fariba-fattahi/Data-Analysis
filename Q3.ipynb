Which non-fossil fuel energy technology will have the best price in the future?

```python
import pandas as pd
import numpy as np
data=pd.read_csv("C:/Users/User/Downloads/levelized-cost-of-energy.csv")
df = pd.DataFrame(data)
```

```python
#Filter the DataFrame for global data ('World') and display all the rows and creating a copy

df_world = df[df['Entity'] == 'World'].copy()

#Rename the columns to improve readability
df_world.rename(columns=lambda x: x.replace(" levelized cost of energy", ""), inplace=True)
display(df_world.sort_values(by='Year'))

```

```python
# Filter the DataFrame for years between 2010 and 2024 (inclusive)
df_filtered = df_world[(df_world['Year'] >= 2010) & (df_world['Year'] <= 2024)]
display(df_filtered)
```

```python
#A separate copy of this filtered dataset was created to prevent changes to the original data 
df_filtered_years= df_filtered.copy()

# Handling missing values through interpolation ensures there is no missing value 
df_filtered_years['Geothermal'] = df_filtered_years['Geothermal'].interpolate(method='linear').copy()

#confrim the result
display(df_filtered_years)
```

```python
# Prepare input (X) and output (y) arrays for each selected energy column to predict future values by performing linear regression
predicted_costs_2027 = {}

# Define columns for regression, excluding 'Year'
regression_target_columns = ['Bioenergy','Geothermal','Offshore wind','Solar photovoltaic','Concentrated solar power','Hydropower','Onshore wind']

for energy in regression_target_columns:
    X = df_filtered_years['Year'].values
    y = df_filtered_years[energy].values

    # Manual calculation of linear regression parameters
    mean_X = np.mean(X)
    mean_y = np.mean(y)

    numerator = np.sum((X - mean_X) * (y - mean_y))
    denominator = np.sum((X - mean_X)**2)

    slope = numerator / denominator
    intercept = mean_y - slope * mean_X

    
    # Predict for the year 2027
    cost_2027 = slope * 2027 + intercept

    predicted_costs_2027[energy] = cost_2027

print("Predicted levelized costs for 2027:")
print(predicted_costs_2027)
```

The previous step successfully predicted the levelized costs for 2030 for each energy technology. However, one of the predictions ('onshore_wind') resulted in a negative value, which is unrealistic for energy costs. Before proceeding to identify the lowest cost and visualize the results, I need to address this anomaly. I will consider zero value for onshore_wind energy column 



```python
# Convert predictions to a DataFrame and correct the unrealistic negative cost for Solar photovoltaic by setting it to zero
predicted_costs_df = pd.DataFrame(list(predicted_costs_2027.items()), columns=['Energy', 'Predicted_Cost_2030'])
predicted_costs_2027['Solar photovoltaic']=0
print(predicted_costs_2027)
```

```python
#for drawing the regression graph, first I add predicted values.
# Create a copy of df_filtered_years to work with
df_filtered_years_copy = df_filtered_years.copy()

# Create a dictionary for the new 2027 row
new_row_2027 = {'Entity': 'World', 'Code': 'OWID_WRL', 'Year': 2027}

# Add predicted costs to the new row
for energy, cost in predicted_costs_2027.items():
    new_row_2027[energy] = cost

# Convert the dictionary to a DataFrame
df_predicted_2027_row = pd.DataFrame([new_row_2027])

# Concatenate the new row with the existing df_world DataFrame
df_world_extended = pd.concat([df_filtered_years_copy, df_predicted_2027_row], ignore_index=True)

df_world_extended
```

```python
import matplotlib.pyplot as plt

# Creating the DataFrame
df = pd.DataFrame(df_world_extended)


columns_to_plot = ['Bioenergy', 'Geothermal', 'Offshore wind', 'Solar photovoltaic', 'Concentrated solar power', 'Hydropower', 'Onshore wind'
]

df = df.set_index('Year')
df[columns_to_plot].plot(
    kind='line', marker='o',
    figsize=(7, 6)
)

plt.title('Fluctuation of Energy Costs (2010 - 2027)', fontsize=10)
plt.xlabel(('Year'), fontsize=10)
plt.ylabel('Cost: US$ per kilowatt-hour', fontsize=10)
plt.grid(True, linestyle='--', alpha=0.9)

# Add a legend to identify the lines
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left' ,fontsize=10)

# Ensure x-axis ticks are integers
plt.gca().xaxis.set_major_locator(plt.MaxNLocator(integer=True))

plt.show()
```
